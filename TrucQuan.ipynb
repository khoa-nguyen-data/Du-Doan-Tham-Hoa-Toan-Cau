```python
# Nh·∫≠p c√°c th∆∞ vi·ªán c·∫ßn thi·∫øt
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# C·∫•u h√¨nh t·ªëi ∆∞u cho m√°y y·∫øu
import plotly.io as pio
pio.templates.default = "plotly_white"
pio.renderers.default = 'notebook'  # Hi·ªÉn th·ªã nh·∫π h∆°n

print("‚úÖ ƒê√£ t·ªëi ∆∞u c·∫•u h√¨nh cho m√°y y·∫øu")
```

```python
# ƒê·ªçc d·ªØ li·ªáu
df = pd.read_csv('global_disaster_response_2018_2024.csv')

# Chuy·ªÉn ƒë·ªïi ƒë·ªãnh d·∫°ng ng√†y
df['date'] = pd.to_datetime(df['date'], format='%d/%m/%Y')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['month_name'] = df['date'].dt.strftime('%B')

# L√†m s·∫°ch d·ªØ li·ªáu - lo·∫°i b·ªè gi√° tr·ªã b·∫•t th∆∞·ªùng
df = df[df['severity_index'] > 0]
df = df[df['casualties'] >= 0]
df = df[df['economic_loss_usd'] >= 0]

# D·ªãch t√™n lo·∫°i th·∫£m h·ªça sang ti·∫øng Vi·ªát
disaster_vn = {
    'Earthquake': 'ƒê·ªông ƒë·∫•t',
    'Flood': 'L≈© l·ª•t',
    'Hurricane': 'B√£o',
    'Wildfire': 'Ch√°y r·ª´ng',
    'Drought': 'H·∫°n h√°n',
    'Tornado': 'L·ªëc xo√°y',
    'Landslide': 'S·∫°t l·ªü ƒë·∫•t',
    'Extreme Heat': 'N·∫Øng n√≥ng c·ª±c ƒëoan',
    'Storm Surge': 'B√£o bi·ªÉn',
    'Volcanic Eruption': 'N√∫i l·ª≠a'
}
df['loai_tham_hoa'] = df['disaster_type'].map(disaster_vn)

# Xem th√¥ng tin d·ªØ li·ªáu
df.head()

print(f"‚úÖ ƒê√£ l√†m s·∫°ch v√† d·ªãch d·ªØ li·ªáu sang ti·∫øng Vi·ªát")
print(f"üí• Lo·∫°i th·∫£m h·ªça: {df['loai_tham_hoa'].nunique()}")
print(f"üåç S·ªë qu·ªëc gia: {df['country'].nunique()}")
print(f"üìÖ Kho·∫£ng th·ªùi gian: {df['date'].min().strftime('%d/%m/%Y')} - {df['date'].max().strftime('%d/%m/%Y')}")
print(f"üìä T·ªïng s·ªë s·ª± ki·ªán: {len(df):,}")
```
