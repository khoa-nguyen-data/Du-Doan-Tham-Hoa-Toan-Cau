```python
import pandas as pd
import plotly.express as px

# Đọc dữ liệu
df = pd.read_csv("global_disaster_response_2018_2024.csv")

# Đếm số sự kiện theo quốc gia
country_count = (
    df.groupby("country")
      .size()
      .reset_index(name="num_disasters")
)
```

```python
fig = px.choropleth(
    country_count,
    locations="country",
    locationmode="country names",   # RẤT QUAN TRỌNG
    color="num_disasters",
    color_continuous_scale="Reds",
    title="World Map of Countries in Disaster Dataset (2018–2024)"
)

fig.show()
```

# chuẩn hóa dữ liệu

```python
# chuẩn hóa tọa độ
df["latitude"] = pd.to_numeric(df["latitude"])
df["longitude"] = pd.to_numeric(df["longitude"])
```

```python
# chuẩn hóa ngày tháng
df["date"] = pd.to_datetime(df["date"], errors="coerce")
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
```

```python
df.to_csv(
    "data_da_tien_xu_ly.csv",
    index=False,
    encoding="utf-8-sig"
)
```
